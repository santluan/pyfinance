```python
import yfinance as yf
import pandas as pd
import warnings
import vectorbt as vbt
import numpy as np
warnings.filterwarnings("ignore")
```

```python
## Coletando dados do IBOVESPA
tickers = ['RRRP3','ALSO3','ALPA4','ABEV3','ARZZ3','ASAI3','AZUL4','B3SA3','BPAN4','BBSE3','BBDC3','BBDC4','BRAP4','BBAS3','BRKM5','BRFS3','BPAC11','CRFB3','CCRO3','CMIG4','CIEL3','COGN3','CPLE6','CSAN3','CPFE3','CMIN3','CVCB3','CYRE3','DXCO3','ECOR3','ELET3','ELET6','EMBR3','ENBR3','ENGI11','ENEV3','EGIE3','EQTL3','EZTC3','FLRY3','GGBR4','GOAU4','GOLL4','NTCO3','SOMA3','HAPV3','HYPE3','IGTI11','ITSA4','ITUB4','JBSS3','KLBN11','RENT3','LWSA3','LREN3','MGLU3','MRFG3','CASH3','BEEF3','MRVE3','MULT3','PCAR3','PETR3','PETR4','PRIO3','PETZ3','QUAL3','RADL3','RAIZ4','RDOR3','RAIL3','SBSP3','SANB11','SMTO3','CSNA3','SLCE3','SUZB3','TAEE11','VIVT3','TIMS3','TOTS3','UGPA3','USIM5','VALE3','VIIA3','VBBR3','WEGE3','YDUQ3']
tickers_sa = [x + ".SA" for x in tickers]
```

```python
dividends = vbt.YFData.download(tickers_sa, start="2018-01-01", end='2022-12-31').get('Dividends')
```

```python
dividends_12M = dividends.rolling('365D').sum()
dividends_12M
```

```python
close_ibov = yf.download(tickers_sa, start="2018-01-01", end='2022-12-31')['Close']
close_ibov
```

```python
dividends_12M.index = dividends_12M.index.date
close_ibov.index = close_ibov.index.date
```

```python
dy_12M = (dividends_12M / close_ibov) * 100
dy_12M
```

```python
dy_12M.index = pd.to_datetime(dy_12M.index, utc=True)
type(dy_12M.index)
```

```python
mediana_dy = pd.DataFrame(dy_12M.median().sort_values(ascending=False))
mediana_dy.head(10)
```

```python
list_tickers = list(mediana_dy.head(10).index)
```

```python
close_tickers = vbt.YFData.download(list_tickers, start='2012-01-01', end='2022-12-31').get('Close')
```

```python
alocacao = pd.DataFrame.vbt.empty_like(close_tickers, fill_value=np.nan)
alocacao.iloc[0] = [0.10,0.10,0.10,0.10,0.10,0.10,0.10,0.10,0.10,0.10]
```

```python
port_dy = vbt.Portfolio.from_orders(
    close=close_tickers,
    size=alocacao,
    size_type='targetpercent',
    group_by=True,
    cash_sharing=True,
    init_cash=None,
    freq='d'
)
```

```python
port_dy.stats(settings = dict(risk_free = 13.75, freq = 'd'))
```

```python
port_dy.plot().show()
```
